```python
dfs = [foxconn['Close'],ase['Close'],sercomm['Close'],accton['Close'],compal['Close'],TSMC['Close'],MED['Close'],WE['Close'],REA['Close'],LARGAN['Close'],ASUS['Close'],ACER['Close'],Quanta['Close'],Inventec['Close']]
df_all = pd.DataFrame(index=foxconn['Close'].index)
```

```python
for i in dfs:
    df_all = df_all.merge(i, how='left',right_index = True,left_index = True)
df_all
```

```python
df_all.columns = ['foxconn','ase','sercomm','accton','compal','TSMC','MED','WE','REA','LARGAN','ASUS','ACER','Quanta','Inventec']
df_all
```

```python
df_all.plot(figsize=(10,8))
```

```python
df_std = np.std(df_all, ddof=1)
df_std.sort_values(ascending=False)
```

```python

df_corr = df_all.corr()
plt.subplots(figsize=(10, 10)) # 設定畫面大小
```

```python
import seaborn as sns
```

```python

sns.heatmap(df_corr, annot=True, vmax=1, square=True, cmap="Blues")
plt.subplots(figsize=(20, 20)) # 設定畫面大小
plt.show()
```
