```python
import pandas_datareader.data as web
```

```python
window=20
```

```python
#鴻海股票資訊
```

```python
HONHAI = web.DataReader(name='2317.TW',data_source='yahoo',start='2020-1-1',end='2021-6-1')
```

```python
rets2=np.log(HONHAI['Close']/HONHAI['Close'].shift(1))
```

```python
rets2.cumsum().apply(np.exp).plot(figsize=(10,6))
```

```python
HONHAI['min']=HONHAI['Close'].rolling(window=window).min()
HONHAI['mean']=HONHAI['Close'].rolling(window=window).mean()
HONHAI['std']=HONHAI['Close'].rolling(window=window).std()
HONHAI['median']=HONHAI['Close'].rolling(window=window).median()
HONHAI['max']=HONHAI['Close'].rolling(window=window).max()
HONHAI.dropna()
```

```python
ax2=HONHAI[['min','mean','max']].iloc[-118:].plot(figsize=(10,6),style=['g--','r--','b--'],lw=0.8)
```

```python
ax=HONHAI[['min','mean','max']].iloc[-118:].plot(figsize=(10,6),style=['g--','r--','b--'],lw=0.8)
HONHAI['Close'].iloc[-118:].plot(ax=ax,lw=2.0)
```

```python
HONHAI['SMA5']=HONHAI['Close'].rolling(window=5).mean()
HONHAI['SMA20']=HONHAI['Close'].rolling(window=20).mean()
HONHAI['SMA60']=HONHAI['Close'].rolling(window=60).mean()
HONHAI[['Close','SMA5','SMA20','SMA60']].plot(figsize=(10,6))
```

```python
ASE = web.DataReader(name='3711.TW',data_source='yahoo',start='2020-1-1',end='2021-6-1')
```

```python
dfs = [HONHAI['Close'],ASE['Close'],SERCOMM['Close']]
```

```python
df_all = pd.DataFrame(index=ASE['Close'].index)
```

```python
for i in dfs:
    df_all = df_all.merge(i, how='left',right_index = True,left_index = True)
df_all
```

```python
df_all.columns = ['HONHAI','ASE','SERCOMM']
df_all
```

```python
df_all.plot(figsize=(10,10))
```

```python
df_std = np.std(df_all, ddof=1)
```

```python
df_std.sort_values(ascending=False)
```

```python
df_corr = df_all.corr()
plt.subplots(figsize=(10, 10)) # 設定畫面大小
```
